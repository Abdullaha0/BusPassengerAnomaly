# <b>Cleaning APC DATA</b>

## Imports and Loading of the APC file

start with checking for nulls in each column

```python
import pandas as pd

apc = pd.read_csv('../data/raw/apc.csv', sep = ';', encoding = 'latin-1')

apc.isnull().sum()
```

```python
columns_to_check = [
    'stationName',                # always empty?
    'stationCode',                # always empty?
    'gpsDeviation',               # always '0 m'?
    'isGpsMapped',                # always False?
    'wheelchairRampAccessCount',  # always 0?
    'isBicycleRackAccessed',      # always False?
    'area',                       # always empty?
    'netstop_latitude',           # always empty?
    'netstop_longitude',          # always empty?
]

for col in columns_to_check:
    print(col + ":", apc[col].unique(), "\n")
```

We see that these columns will not help in the analysis as some of them are empty, some contain the same boolean value on all rows which does not add any benefit.

# Cleaning

```python
columns_to_drop = [
    'fileName',                   # Doesn't help
    'stationName',                # always empty
    'stationCode',                # always empty
    'gpsDeviation',               # always '0 m'
    'isGpsMapped',                # always False
    'wheelchairRampAccessCount',  # always 0
    'isBicycleRackAccessed',      # always False
    'area',                       # always empty
    'netstop_latitude',           # always empty
    'netstop_longitude',          # always empty
]

apc_clean = apc.drop(columns= columns_to_drop)
apc_clean = apc_clean.drop_duplicates().reset_index(drop=True)

print("before", apc.shape)
print("after", apc_clean.shape)

apc_clean.to_csv('../data/processed/apc_clean.csv', sep=';', index=False, encoding='latin-1')
```

We drop the 10 columns identified above as either always empty or constant. We also drop the 2 fully duplicated rows we found in the EDA. The original `apc` is kept untouched; we work on `apc_clean` from here.
