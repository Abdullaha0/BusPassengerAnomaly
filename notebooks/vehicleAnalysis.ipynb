# <b>Vehicle Analysis</b>
The Dilax API data (matched_stops and matched_trips) only covers 23 vehicles, 
while the raw APC data covers 176 vehicles and Thoreb covers 170. This notebook 
investigates the fleet overlap to decide how to scope the anomaly detection.

**Questions:**
1. Are the 23 vehicles in matched_stops the same as the 23 in matched_trips?
2. How many of those 23 vehicles also appear in apc and thoreb?
3. Should we filter the analysis to the 23-vehicle subset, or run two parallel 
   analyses (full APC fleet + 23-vehicle subset with Dilax corrections)?

## Setup: Imports And Loading Data

```python
import pandas as pd

apc = pd.read_csv('../data/raw/apc.csv', sep = ';', encoding = 'latin-1')
matched_stops = pd.read_csv('../data/raw/matched_stops.csv', sep = ';', decimal=',', encoding = 'latin-1')
matched_trips = pd.read_csv('../data/raw/matched_trips.csv', sep = ';', decimal=',', encoding = 'latin-1')
thoreb = pd.read_csv('../data/raw/thoreb.csv', encoding = 'latin-1', low_memory = False)
```

## Question 1 - Are the 23 vehicles in matched_stops the same as in matched_trips?

```python
stops_vehicles = set(matched_stops['vehicleNumber'].unique())
trips_vehicles = set(matched_trips['data.vehicles.vehicleNumber'].unique())

print(f"matched_stops vehicles: {len(stops_vehicles)}")
print(f"matched_trips vehicles: {len(trips_vehicles)}")
print(f"In both:                {len(stops_vehicles & trips_vehicles)}")
```

**Result:** The same 23 vehicles appear in both files. matched_stops and matched_trips are consistent.
